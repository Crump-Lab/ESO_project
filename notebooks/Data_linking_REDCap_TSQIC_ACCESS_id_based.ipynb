# TSQIC tabs merging

```python
# Load the Excel file
file_path = "Complications_TSQIC_TC_13_02_2025.xlsx" 
df_tsqic = pd.read_excel(file_path, sheet_name=None)

# Display sheet names
print(df_tsqic.keys())

# Access specific sheets/tabs
df_tsqic_1 = df_tsqic["TSQIC-Complications"]  
df_tsqic_2 = df_tsqic["TSQIC-admission-discharge dates"] 
print(df_tsqic_1)
print(df_tsqic_2)
```

```python
# Standardize column names for merging
df_tsqic_1 = df_tsqic_1.rename(columns={"date (operation)": "operation_date", "Readmission":"readmission_30d"})
df_tsqic_2 = df_tsqic_2.rename(columns={"Date (operation)": "operation_date"})


# Convert operation_date to datetime for proper matching
df_tsqic_1["operation_date"] = pd.to_datetime(df_tsqic_1["operation_date"])
df_tsqic_2["operation_date"] = pd.to_datetime(df_tsqic_2["operation_date"])

# Merge with cartesian product for matched rows
merged_matched = df_tsqic_1.merge(df_tsqic_2, on=["id", "operation_date"], how="outer", indicator=True)

# Display the result -> _merge column=> left_only: appear TSQIC-Complications tab only + right_only: appear TSQIC-admission-discharge dates tab only 
print(merged_matched)
```

### Merging examination

```python
# Confirm if merging went well

# IDs in df_tsqic_1 but not in df_tsqic_2 -> no missing value so it makes sense that we dont have "left_only"
missing_in_2 = df_tsqic_1[~df_tsqic_1["id"].isin(df_tsqic_2["id"])]
print("IDs missing in df_tsqic_2:", missing_in_2)

# IDs in df_tsqic_2 but not in df_tsqic_1
missing_in_1 = df_tsqic_2[~df_tsqic_2["id"].isin(df_tsqic_1["id"])]
print("\n\nIDs missing in df_tsqic_1:", missing_in_1)
```

```python
# Go deeper into right_only
right_only_count = (merged_matched["_merge"] == "right_only").sum()
print(f"Number of right-only rows: {right_only_count}")
```

```python
# Convert date columns to strings if they are datetime (to avoid potential type mismatches)
df_tsqic_1["date"] = df_tsqic_1["operation_date"].astype(str)
df_tsqic_2["date"] = df_tsqic_2["operation_date"].astype(str)

# Create a set of (ID, Date) pairs from df_tsqic_1
id_date_set_1 = set(zip(df_tsqic_1["id"], df_tsqic_1["date"]))

# Filter rows in df_tsqic_2 where (ID, Date) is not in df_tsqic_1
missing_pairs_in_1 = df_tsqic_2[~df_tsqic_2.apply(lambda row: (row["id"], row["date"]) in id_date_set_1, axis=1)]

print("ID-Date pairs in df_tsqic_2 but not in df_tsqic_1:\n", missing_pairs_in_1[["id", "date"]])
#677 rows matched! so it went well. 
```

```python
#estimated_los: difference between these two dates most often corresponds with length of stay, LOS
merged_matched["estimated_los"] = (merged_matched["DischargeDate"] - merged_matched["operation_date"]).dt.days
print(merged_matched["estimated_los"].describe())  # Summary statistics
print("\n\n")
print(merged_matched[merged_matched["estimated_los"] < 0])  # Check for negative LOS values
print("\n\n")
print(merged_matched[merged_matched["estimated_los"].isna()])  # Check for missing values
```

```python
merged_matched
```

```python
merged_matched["_merge"] = merged_matched["_merge"].replace("right_only", "secondtab_only")
merged_matched
```

```python
merged_matched = merged_matched.rename(columns={"_merge": "_merge_tsqic"})
merged_matched
```

```python
# Save to a new Excel file
merged_matched.to_excel("Merged_TSQIC.xlsx", index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Function to analyze LOS differences
def analyze_los_differences(data):
    """
    Analyze differences between three length of stay variables:
    los, estimated_los_access, and estimated_los_tsqic
    
    Parameters:
    data (DataFrame): DataFrame containing the LOS variables
    
    Returns:
    dict: Dictionary containing summary statistics and processed data
    """
    # Create a copy to avoid modifying the original data
    df = data.copy()
    
    # Check NA counts for each column
    na_summary = pd.DataFrame({
        'los_na': [df['los'].isna().sum()],
        'estimated_los_access_na': [df['estimated_los_access'].isna().sum()],
        'estimated_los_tsqic_na': [df['estimated_los_tsqic'].isna().sum()],
        'total_rows': [len(df)]
    })
    
    # Check if columns are date type and convert to numeric days if needed
    for col in ['los', 'estimated_los_access', 'estimated_los_tsqic']:
            df[f'{col}_days'] = df[col]
    
    # Calculate absolute differences between pairs
    df['diff_los_access'] = np.abs(df['los_days'] - df['estimated_los_access_days'])
    df['diff_los_tsqic'] = np.abs(df['los_days'] - df['estimated_los_tsqic_days'])
    df['diff_access_tsqic'] = np.abs(df['estimated_los_access_days'] - df['estimated_los_tsqic_days'])
    
    # Calculate summary statistics
    diff_summary = pd.DataFrame({
        'mean_diff_los_access': [df['diff_los_access'].mean()],
        'median_diff_los_access': [df['diff_los_access'].median()],
        'max_diff_los_access': [df['diff_los_access'].max()],
        
        'mean_diff_los_tsqic': [df['diff_los_tsqic'].mean()],
        'median_diff_los_tsqic': [df['diff_los_tsqic'].median()],
        'max_diff_los_tsqic': [df['diff_los_tsqic'].max()],
        
        'mean_diff_access_tsqic': [df['diff_access_tsqic'].mean()],
        'median_diff_access_tsqic': [df['diff_access_tsqic'].median()],
        'max_diff_access_tsqic': [df['diff_access_tsqic'].max()],
        
        # Count exact matches
        'exact_match_los_access': [(df['diff_los_access'] == 0).sum()],
        'exact_match_los_tsqic': [(df['diff_los_tsqic'] == 0).sum()],
        'exact_match_access_tsqic': [(df['diff_access_tsqic'] == 0).sum()],
        
        # Count differences within 1 day
        'within_1day_los_access': [(df['diff_los_access'] <= 1).sum()],
        'within_1day_los_tsqic': [(df['diff_los_tsqic'] <= 1).sum()],
        'within_1day_access_tsqic': [(df['diff_access_tsqic'] <= 1).sum()]
    })
    
    # Create a dataset with complete cases for visualization
    complete_cases = df.dropna(subset=['los_days', 'estimated_los_access_days', 'estimated_los_tsqic_days'])
    
    return {
        'na_summary': na_summary,
        'diff_summary': diff_summary,
        'complete_cases': complete_cases,
        'all_data': df
    }

# Function to create visualizations
def create_los_plots(analysis_results):
    """
    Create visualizations for LOS comparison
    
    Parameters:
    analysis_results (dict): Output from analyze_los_differences function
    
    Returns:
    dict: Dictionary containing the plot figures
    """
    complete_cases = analysis_results['complete_cases']
    
    # Set up the figure layout
    fig1, axes1 = plt.subplots(1, 3, figsize=(18, 6))
    
    # Scatter plot: LOS vs Estimated LOS (ACCESS)
    axes1[0].scatter(complete_cases['los_days'], complete_cases['estimated_los_access_days'], alpha=0.3)
    axes1[0].plot([0, complete_cases[['los_days', 'estimated_los_access_days']].max().max()], 
                 [0, complete_cases[['los_days', 'estimated_los_access_days']].max().max()], 
                 'r-')
    axes1[0].set_title('LOS vs Estimated LOS (ACCESS)')
    axes1[0].set_xlabel('Actual LOS (days)')
    axes1[0].set_ylabel('Estimated LOS ACCESS (days)')
    
    # Scatter plot: LOS vs Estimated LOS (TSQIC)
    axes1[1].scatter(complete_cases['los_days'], complete_cases['estimated_los_tsqic_days'], alpha=0.3)
    axes1[1].plot([0, complete_cases[['los_days', 'estimated_los_tsqic_days']].max().max()], 
                 [0, complete_cases[['los_days', 'estimated_los_tsqic_days']].max().max()], 
                 'r-')
    axes1[1].set_title('LOS vs Estimated LOS (TSQIC)')
    axes1[1].set_xlabel('Actual LOS (days)')
    axes1[1].set_ylabel('Estimated LOS TSQIC (days)')
    
    # Scatter plot: Estimated LOS (ACCESS) vs Estimated LOS (TSQIC)
    axes1[2].scatter(complete_cases['estimated_los_access_days'], complete_cases['estimated_los_tsqic_days'], alpha=0.3)
    axes1[2].plot([0, complete_cases[['estimated_los_access_days', 'estimated_los_tsqic_days']].max().max()], 
                 [0, complete_cases[['estimated_los_access_days', 'estimated_los_tsqic_days']].max().max()], 
                 'r-')
    axes1[2].set_title('Estimated LOS (ACCESS) vs Estimated LOS (TSQIC)')
    axes1[2].set_xlabel('Estimated LOS ACCESS (days)')
    axes1[2].set_ylabel('Estimated LOS TSQIC (days)')
    
    plt.tight_layout()
    
    # Histograms of differences
    fig2, axes2 = plt.subplots(1, 3, figsize=(18, 6))
    
    # Histogram: LOS vs ACCESS
    sns.histplot(analysis_results['all_data']['diff_los_access'].dropna(), bins=30, kde=True, ax=axes2[0])
    axes2[0].set_title('Difference: LOS vs Estimated LOS (ACCESS)')
    axes2[0].set_xlabel('Absolute Difference (days)')
    
    # Histogram: LOS vs TSQIC
    sns.histplot(analysis_results['all_data']['diff_los_tsqic'].dropna(), bins=30, kde=True, ax=axes2[1])
    axes2[1].set_title('Difference: LOS vs Estimated LOS (TSQIC)')
    axes2[1].set_xlabel('Absolute Difference (days)')
    
    # Histogram: ACCESS vs TSQIC
    sns.histplot(analysis_results['all_data']['diff_access_tsqic'].dropna(), bins=30, kde=True, ax=axes2[2])
    axes2[2].set_title('Difference: Estimated LOS (ACCESS) vs (TSQIC)')
    axes2[2].set_xlabel('Absolute Difference (days)')
    
    plt.tight_layout()
    
    return {'comparison_plots': fig1, 'difference_plots': fig2}

# Function to identify outliers
def identify_outliers(data, threshold_days=7):
    """
    Identify outliers where differences exceed a threshold
    
    Parameters:
    data (DataFrame): DataFrame with calculated differences
    threshold_days (int): Threshold in days for considering a difference as an outlier
    
    Returns:
    DataFrame: Subset of data containing outliers
    """
    outliers = data[
        (data['diff_los_access'] > threshold_days) | 
        (data['diff_los_tsqic'] > threshold_days) | 
        (data['diff_access_tsqic'] > threshold_days)
    ]
    
    return outliers



# Running the analysis
analysis_results = analyze_los_differences(merged_final_tsqic_redcap_access)

# Print summaries
print("NA Summary:")
print(analysis_results['na_summary'])
print("\nDifference Summary:")
print(analysis_results['diff_summary'])

# Create and show plots
plots = create_los_plots(analysis_results)
plt.show()

# Find outliers
outliers = identify_outliers(analysis_results['all_data'])
print(f"\nFound {len(outliers)} outliers with differences > 7 days")
print(outliers[['los', 'estimated_los_access', 'estimated_los_tsqic', 
                'diff_los_access', 'diff_los_tsqic', 'diff_access_tsqic']].head())


# Additional analysis: Percentage agreement
def calculate_agreement_stats(analysis_results):
    """
    Calculate agreement statistics between the three LOS measures
    
    Parameters:
    analysis_results (dict): Output from analyze_los_differences function
    
    Returns:
    DataFrame: Agreement statistics
    """
    all_data = analysis_results['all_data']
    total_valid = len(all_data.dropna(subset=['los_days', 'estimated_los_access_days', 'estimated_los_tsqic_days']))
    
    # Calculate percentages for different thresholds
    agreement_stats = pd.DataFrame({
        'comparison': ['LOS vs ACCESS', 'LOS vs TSQIC', 'ACCESS vs TSQIC'],
        'exact_match': [
            (all_data['diff_los_access'] == 0).sum() / all_data['diff_los_access'].count() * 100,
            (all_data['diff_los_tsqic'] == 0).sum() / all_data['diff_los_tsqic'].count() * 100,
            (all_data['diff_access_tsqic'] == 0).sum() / all_data['diff_access_tsqic'].count() * 100
        ],
        'within_1_day': [
            (all_data['diff_los_access'] <= 1).sum() / all_data['diff_los_access'].count() * 100,
            (all_data['diff_los_tsqic'] <= 1).sum() / all_data['diff_los_tsqic'].count() * 100,
            (all_data['diff_access_tsqic'] <= 1).sum() / all_data['diff_access_tsqic'].count() * 100
        ],
        'within_3_days': [
            (all_data['diff_los_access'] <= 3).sum() / all_data['diff_los_access'].count() * 100,
            (all_data['diff_los_tsqic'] <= 3).sum() / all_data['diff_los_tsqic'].count() * 100,
            (all_data['diff_access_tsqic'] <= 3).sum() / all_data['diff_access_tsqic'].count() * 100
        ],
        'within_7_days': [
            (all_data['diff_los_access'] <= 7).sum() / all_data['diff_los_access'].count() * 100,
            (all_data['diff_los_tsqic'] <= 7).sum() / all_data['diff_los_tsqic'].count() * 100,
            (all_data['diff_access_tsqic'] <= 7).sum() / all_data['diff_access_tsqic'].count() * 100
        ]
    })
    
    return agreement_stats


# Calculate agreement statistics
agreement_stats = calculate_agreement_stats(analysis_results)
print("\nAgreement Statistics (%):")
print(agreement_stats)
```
